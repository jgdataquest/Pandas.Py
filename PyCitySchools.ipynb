```python
import pandas as pd
import numpy as np
```

District Summary

```python
school="raw_data/schools_complete.csv"
school_pd = pd.read_csv(school)
school_pd.rename(columns = {'name':'school'}, inplace = True)
students="raw_data/students_complete.csv"
students_pd = pd.read_csv(students)
df = pd.merge(school_pd, students_pd, on=("school"))
total_school=school_pd.count().max()
total_students=students_pd['name'].count()
total_budget=school_pd['budget'].sum()
#Average Math Score
avg_math_score = df['math_score'].sum()/total_students
avg_math_score
#Average Reading Score
avg_reading_score = df['reading_score'].sum()/total_students
avg_reading_score
#% Passing Math
passing_math_percentage =df[df.math_score > 70].count().math_score * 100/total_students
#df[cols] > 0).sum(1)                       
passing_math_percentage
#% Passing Reading
passing_reading_percentage =df[df.reading_score > 70].count().reading_score * 100/total_students
passing_reading_percentage
#Overall Passing Rate (Average of the above two)
#overall_passing_rate = (df[df.reading_score > 70].count().reading_score +\
#                        df[df.math_score > 70].count().math_score) * 100/(2*total_students)
overall_passing_rate = (passing_reading_percentage + passing_math_percentage)/2
overall_passing_rate
# Create a new table consolodating above calculations
dist_summary = pd.DataFrame({"Total Schools": [total_school],
                                   "Total Students": [total_students],
                                   "Total Budget": [total_budget],
                                   "Average Math Score": [avg_math_score],
                                   "Average Reading Score": [avg_reading_score],
                                   "% Passing Math":[passing_math_percentage],
                                   "% Passing Reading":[passing_reading_percentage],
                                   "Overall Passing Rate": [overall_passing_rate]
                            })
dist_summary = dist_summary [["Total Schools","Total Students","Total Budget","Average Math Score",\
                             "Average Reading Score","% Passing Math","% Passing Reading",\
                             "Overall Passing Rate"]]
dist_summary["Total Budget"] = dist_summary["Total Budget"].map("${0:,.0f}".format)
dist_summary
#school_pd.head()
```

School Summary

```python
## School Summary
school_pd['Per Student Budget']=school_pd['budget'] / school_pd['size']
# Using school as Index
indexed_school_pd = school_pd.set_index(['school'])
indexed_school_pd['Average Math Score']=students_pd.groupby('school')['math_score'].mean()
indexed_school_pd['Average Reading Score']=students_pd.groupby('school')['reading_score'].mean()
#% Passing Math
math_passing_count=students_pd[students_pd.math_score > 70].groupby('school')['math_score'].count() 
indexed_school_pd['% Passing Math']= math_passing_count * 100/indexed_school_pd['size']
#% Passing Reading
reading_passing_count=students_pd[students_pd.reading_score > 70].groupby('school')['reading_score'].count() 
indexed_school_pd["% Passing reading"]= reading_passing_count * 100/indexed_school_pd['size']
#Overall Passing Rate (Average of the above two)
indexed_school_pd['% Overall Passing Rate'] = \
           (indexed_school_pd['% Passing Math'] + indexed_school_pd['% Passing reading'])/2
del indexed_school_pd['School ID']
indexed_school_pd.rename(columns = {'type':'School Type', 'size':'Total Students','budget':'Total School Budget'}, inplace = True)
school_summary_pd=indexed_school_pd.reset_index()
school_summary_pd
```

Top 5 Performing Schools (By Passing Rate)

```python
indexed_school_pd.sort_values(["% Overall Passing Rate"], ascending=False, inplace=True)
indexed_school_pd.head()
```

Math Scores by Grade

```python
#Math Scores by Grade
math_score_grade=pd.pivot_table(students_pd, index=["school"], columns= "grade", values="math_score", aggfunc = 'mean')
math_score_grade = math_score_grade[['9th', '10th', '11th', '12th']]
math_score_grade
```

Reading Scores by Grade

```python
# Reading Scores by Grade
Reading_score_grade=pd.pivot_table(students_pd, index=["school"], columns= "grade", values="reading_score", aggfunc = 'mean')
Reading_score_grade = Reading_score_grade[['9th', '10th', '11th', '12th']]
Reading_score_grade
```

Scores by School Spending

```python
#Scores by School Spending
indexed_school_pd['spending_Range']=pd.cut(indexed_school_pd['Per Student Budget'],[0,585,615,645,675],\
                                    right=False, labels=['<$585','$585-615','$615-645','$645-675'])
spending_range_grade=indexed_school_pd.groupby('spending_Range')['Average Math Score','Average Reading Score','% Passing Math','% Passing reading','% Overall Passing Rate'].\
                                       agg({'Average Math Score':'mean','Average Reading Score':'mean','% Passing Math':'mean','% Passing reading':'mean','% Overall Passing Rate':'mean'})                                 
spending_range_grade.reindex(['<$585','$585-615','$615-645','$645-675'])
```
